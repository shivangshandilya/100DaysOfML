```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
df.head(10)
```

```python
df.shape
```

```python
# cleaning the data, we're looking for data of USA races, 50km ones and in 2020
```

```python
df[(df["Event distance/length"].isin(["50km", "50mi"]))  & (df["Year of event"] == 2020)]
```

```python
df[df["Event name"] == "Everglades 50 Mile Ultra Run (USA)"]
```

```python
df[df["Event name"].str.split('(').str.get(1).str.split(')').str.get(0) == "USA"]
```

```python
# combine all the filters

df[
    (df["Event distance/length"].isin(["50km", "50mi"])) &
    (df["Year of event"] == 2020) &
    (df["Event name"].str.split('(').str.get(1).str.split(')').str.get(0) == "USA")
]
```

```python
df2 = df[
    (df["Event distance/length"].isin(["50km", "50mi"])) &
    (df["Year of event"] == 2020) &
    (df["Event name"].str.split('(').str.get(1).str.split(')').str.get(0) == "USA")
]
```

```python
df2.head(10)
```

```python
# remove USA from event name

df2["Event name"] = df2["Event name"].str.split('(').str.get(0)
```

```python
df2.head(10)
```

```python
# clean up athelete's age

df2["athlete_age"] = 2020 - df2["Athlete year of birth"]
```

```python
# remove h from athlete performance

df2["Athlete performance"] = df2["Athlete performance"].str.split(' ').str.get(0)
```

```python
df2.head()
```

```python
# drop columns: Athlete Club, Athlete Country, Athelete YOB, Athlete Age Category

df2 = df2.drop(["Athlete club", "Athlete country", "Athlete year of birth", "Athlete age category"], axis = 1)
```

```python
df2.head()
```

```python
# clean up null values

df2.isna().sum()
```

```python
df2[df2["athlete_age"].isnull() == 1]
```

```python
df2 = df2.dropna()
```

```python
# check for dupes

df2.duplicated().sum() # no dupes
```

```python
# reset index

df2.reset_index(drop = True)
```

```python
# fix types

df2.dtypes
```

```python
df2["athlete_age"] = df2["athlete_age"].astype(int)
```

```python
df2["Athlete average speed"] = df2["Athlete average speed"].astype(float)
```

```python
df2.dtypes
```

```python
# rename cols

df2 = df2.rename(columns = {
                            'Year of event': 'year',                  
                            'Event dates': 'race_day',                  
                            'Event name': 'race_name',                    
                            'Event distance/length': 'race_length',         
                            'Event number of finishers': 'race_no_of_finishers',      
                            'Athlete performance': 'athlete_performance',           
                            'Athlete gender': 'athlete_gender',                
                            'Athlete average speed': 'athlete_average_speed',        
                            'Athlete ID': 'athlete_id'                         
})
```

```python
df2.head()
```

```python
# reorder cols

df3 = df2[["race_day","race_name","race_length","race_no_of_finishers","athlete_performance","athlete_gender","athlete_average_speed","athlete_id","athlete_age", "year"]]
df3.head()
```
