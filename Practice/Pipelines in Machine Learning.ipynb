```python
from sklearn.pipeline import Pipeline

# feature scaling
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
steps = [("standard_scaler", StandardScaler()),
        ("Classifier", LogisticRegression())]
```

```python
steps
```

```python
Pipeline(steps)
```

```python
pipe = Pipeline(steps)
```

```python
# lets create our own dataset
from sklearn.datasets import make_classification
X, y = make_classification(n_samples = 1000)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
X_train
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

## EXAMPLE- 2
Displaying a pipeline with standard scaler, dimensionality reduction and then estimator

```python
from sklearn.decomposition import PCA
from sklearn.svm import SVC
```

```python
steps = [("scaling", StandardScaler()),
        ("PCA", PCA(n_components = 3)),
        ("SVC", SVC())]
```

```python
Pipeline(steps)
```

```python
pipe2 = Pipeline(steps)
```

```python
pipe2.fit(X_train, y_train)
```

```python
pipe2.predict(X_test)
```

## EXAMPLE-3
Complex ex of columns transformer

```python
from sklearn.impute import SimpleImputer
```

```python
# numerical processing pipeline
import numpy as np
numeric_processor = Pipeline(
    steps = [("imputation_mean", SimpleImputer(missing_values = np.nan, strategy = "mean")),
            ("scaler", StandardScaler())]
)
```

```python
numeric_processor
```

```python
# categorical processing pipeline
from sklearn.preprocessing import OneHotEncoder
categorical_processor = Pipeline(
    steps = [("imputation_constant", SimpleImputer(fill_value = "missing", strategy = "constant")),
            ("onehot", OneHotEncoder(handle_unknown = "ignore"))]
)
```

```python
categorical_processor
```

```python
# combining processing techniques
from sklearn.compose import ColumnTransformer
```

```python
preprocessor = ColumnTransformer(
    [("categorical", categorical_processor, ["gender", "City"]),
    ("numerical", numeric_processor, ["age", "height"])]
)
```

```python
preprocessor
```

```python
from sklearn.pipeline import make_pipeline
```

```python
make_pipeline(preprocessor, LogisticRegression())
```
