# ReLU (Rectified Linear Unit)

**used in Neural networks / MLPs, between layers to introduce non-linearity.
It allows networks to learn complex patterns, not just straight lines or hyperplanes.**

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
  return np.maximum(x,0)

x=np.linspace(-10,10)
y=relu(x)
print(y)
plt.title("ReLU")
plt.plot(x,y)
```

# Leaky ReLU

**Leaky ReLU is a variant of the standard ReLU activation function that allows a small, non-zero gradient when the input is negative.**

This means gradients are zero for negative values.

Neurons can "die" during training (stop updating completely).

**Instead of a flat zero slope on the negative side, Leaky ReLU gives a small slope. This allows some gradient flow, even for negative inputs.**


```python
def leakyrelu(x):
  return np.maximum(x,0.1*x)

x=np.linspace(-10,10)
y=leakyrelu(x)
print(y)
plt.title("Leaky ReLU")
plt.plot(x,y)
```

# PReLU (Parametric ReLU)

**PReLU is an advanced version of Leaky ReLU, where the slope for negative inputs is learned during training rather than being a fixed small value like 0.01.**

```python
def prelu(x,alpha):
  return np.maximum(x,alpha*x)

x=np.linspace(-10,10)
y=prelu(x,0.5)
print(y)
plt.title("PReLU (alpha=.5)")
plt.plot(x,y)
```

# ELU (Exponential Linear Unit)

**ELU is another activation function designed to fix some of the problems in ReLU, especially the "dying ReLU" issue, while providing smoother outputs.**

ùõº is a small positive number (default: 1.0).

For positive values, ELU behaves just like ReLU.

For negative values, it smoothly tapers off instead of going flat or linear.

```python
def elu(x,alpha):
  return np.where(x >= 0,x,alpha*(np.exp(x)-1))

x=np.linspace(-10,10)
y=elu(x,0.2)
print(y)
plt.title("eLU")
plt.plot(x,y)
```
